# Data Visualization

## Univariate Analysis: It means analysing single variable through Data visualization

```python
plt.figure(figsize=(10,6),facecolor='pink')
plt.hist(data['Sales'],color='purple',edgecolor = "black")
plt.title('Sales_Analysing')
plt.xlabel('Sales')
plt.ylabel('Counts')
plt.show()
#We can say that most of sales is in 10 rs to 15 rs and it is right skewed
```

```python
sns.violinplot(data['TV'],color='y')
plt.title("Analysing Tv column")
plt.axvline(data['TV'].median())
plt.show()
#we can infer that 50% of data if less than 150 and other 50% of data is more than 150
```

```python
plt.figure(figsize=(10,6))
plt.hist(data['Newspaper'],color='brown',edgecolor = "black")
plt.title('Visualizing Newspaper')
plt.xlabel('Newspaper')
plt.ylabel('Frequency')
plt.show()
# Here we can infer that mostly newpaper value is 0-20
```

```python
sns.boxplot(data['Radio'])
plt.title('Analysing Radio values')
plt.show()
#it states that values of radio lies between 0 to 50
```

## Multi variate Analysing : It means analysing two variables or more variables together

```python
#Here,we are analysing Tv and sales
plt.figure(figsize=(10,6))
sns.regplot(y=data['TV'],x=data['Sales'],color='g')
plt.title("Relation between Tv and Sales")
plt.show()
#we can see their is positive relation between tv and sales 
```

```python
#Here,we are analysing Radio and sales
plt.figure(figsize=(10,6))
plt.scatter(data['Radio'],data['Sales'],color='pink',edgecolors='black',s=100)
plt.title("Relation between Radio and Sales")
plt.xlabel('Radio')
plt.ylabel('Sales')
plt.grid()
plt.show()
#we can see their is positive relationship between Radio and sales but ascompare to tv it is slightly low positive relation
```

```python
#Here,we are analysing Tv and sales
plt.figure(figsize=(10,6))
sns.regplot(y=data['Newspaper'],x=data['Sales'],color='r')
plt.title("Relation between Newspaper and Sales")
plt.show()
#we can see their is very very low positive relationship between Newspaper and sales 
```

```python
# Visualizing coorelation through heatmap
plt.figure(figsize=(10,6))
sns.heatmap(d.corr(),annot=True)
plt.title("Relation between all numerical columns")
plt.show()
```
